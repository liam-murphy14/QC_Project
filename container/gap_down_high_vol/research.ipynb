```python
# QuantBook Analysis Tool 
# For more information see [https://www.quantconnect.com/docs/research/overview]
qb = QuantBook()
upst = qb.AddEquity('UPST', Resolution.Minute).Symbol
ozon = qb.AddEquity('OZON', Resolution.Minute).Symbol
yndx = qb.AddEquity('YNDX', Resolution.Minute).Symbol
hhr = qb.AddEquity('HHR', Resolution.Minute).Symbol
```

```python
hist = qb.History([upst, ozon, yndx, hhr], timedelta(days=200), Resolution.Minute)
hist
```

```python
%matplotlib inline
import datetime
```

```python
bad_datetime = datetime.datetime(2022, 5, 10)
after_bad_day = hist.loc[upst].index >= bad_datetime - timedelta(days=1)
before_bad_day = hist.loc[upst].index <= bad_datetime + timedelta(days=1)
bad_day = after_bad_day * before_bad_day
hist.loc[upst][bad_day]['askhigh'].plot()
```

```python
for symb in hist.index.levels[0]:
    hist.loc[symb].drop('volume', axis=1).plot()
```
